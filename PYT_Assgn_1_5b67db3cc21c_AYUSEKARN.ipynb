# Case 4: Presentase Kejahatan terjadi London 2010-2016

```python
df_region2= df_crime.groupby('region')['year'].sum()
df_region2
```

```python
df_region2.plot(
    kind="pie",
    figsize = (15,6),
    startangle = 90,
    autopct="%1.1f%%",
    shadow=True,
    pctdistance=1.12,
    labels=None
)
plt.title("Presentase kejahatan di London 2010-2016", y=1.12)
plt.axis ('equal')
plt.legend(labels=df_region2.index, loc="upper left")
plt.show()
```

Presentase kejahatan yang terjadi di seluruh wilayah london pada tahun 2010-2016

```python
! pip install wordcloud
from wordcloud import WordCloud, STOPWORDS
print("WordCloud is installed and imported")
```

```python
df_Wcd = df_crime[df_crime['year'].isin([2010,2011,2012,2013,2014,2015,2016])]
df_Wcd = df_Wcd[df_Wcd['region'].isin(['City of London'])]
df_Wcd = df_Wcd['major_category'].str.cat(sep='')

stopwords = set(STOPWORDS)
df_wordcloud = WordCloud(
    background_color='pink',
    max_words=2000,
    stopwords=stopwords
)
df_wordcloud.generate(df_Wcd)
plt.imshow(df_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

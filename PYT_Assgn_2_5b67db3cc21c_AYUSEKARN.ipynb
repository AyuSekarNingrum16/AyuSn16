```python
nyc = pd.read_csv('nyc-rolling-sales.csv')
print (nyc)
```

```python
nyc.drop(['Unnamed: 0', 'EASE-MENT'], axis=1, inplace=True)
nyc.head()
```

```python
nyc.shape
```

```python
nyc.info()
```

```python
nyc['Borough1'] = nyc['BOROUGH'].replace ({1:'Manhattan', 2 :'Brooklyn',3: 'Queens', 4:'Bronx', 5:'Stated Island'})
nyc.head()
```

```python
print('Data Name')
for col in nyc.columns :
    if nyc[col].dtype =='object' :
        print(col,nyc[col].nunique())
```

```python
numer = ['LAND SQUARE FEET', 'GROSS SQUARE FEET', 'SALE PRICE', 'BOROUGH']
for col in numer :
    nyc[col] = pd.to_numeric(nyc[col], errors='coerce')
categ = ['NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT',
'BUILDING CLASS AT TIME OF SALE', 'TAX CLASS AT TIME OF SALE']
for col in categ : 
    nyc[col] = nyc[col].astype('category')

nyc['SALE DATE'] = pd.to_datetime(nyc['SALE DATE'], errors = 'coerce')
```

```python
missing = nyc.isnull().sum()/len(nyc)*100

print (pd.DataFrame([missing[missing>0], pd.Series(nyc.isnull().
sum()[nyc.isnull().sum()>1000])], index =['percent missing', 'how many missing']))
```

```python
print(nyc[(nyc['LAND SQUARE FEET'].isnull()) & (nyc['GROSS SQUARE FEET'].notnull())].shape)
print(nyc[(nyc['LAND SQUARE FEET'].notnull()) & (nyc['GROSS SQUARE FEET'].isnull())].shape)
```

```python
nyc['LAND SQUARE FEET'] = nyc['LAND SQUARE FEET'].mask((nyc['LAND SQUARE FEET'].isnull()) & (nyc['GROSS SQUARE FEET']. notnull()), nyc['GROSS SQUARE FEET'])
nyc['GROSS SQUARE FEET'] = nyc['GROSS SQUARE FEET'].mask((nyc['LAND SQUARE FEET'].notnull()) & (nyc['GROSS SQUARE FEET'].isnull()), nyc['LAND SQUARE FEET'])
```

```python
p_bronx = 4067/(384+4067+1570+171+1+3+2+23)
print("%bronx:", p_bronx)

n = (384+4067+1570+171+1+3+2+23)
print("n : ", n )

se_bronx = np.sqrt(p_bronx * (1- p_bronx) / n)
print("SE :", se_bronx)

z_score = 1.96
moe_bronx = z_score * se_bronx
print("Margin of Error:", moe_bronx)

CI = (p_bronx - moe_bronx, p_bronx + moe_bronx)
print("Confidance Intervals:", CI)
```
